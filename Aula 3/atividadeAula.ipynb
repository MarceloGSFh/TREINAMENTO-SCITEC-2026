```python
df['victory_status'] = df['victory_status'].astype('category')
df['winner'] = df['winner'].astype('category')
df.info(memory_usage='deep')
```

```python
df = df.drop(columns=['white_id', 'black_id', 'moves'])
df['opening_ply'] = df['opening_ply'].astype('int8')
df['white_rating'] = df['white_rating'].astype('int16')
df['black_rating'] = df['black_rating'].astype('int16')
df['turns'] = df['turns'].astype('int16')
df['opening_name'] = df['opening_name'].astype('category')
df['opening_eco'] = df['opening_eco'].astype('category')
df['created_at'] = pd.to_datetime(df['created_at']/1000, unit='s')
df['last_move_at'] = pd.to_datetime(df['last_move_at']/1000, unit='s')

df.info(memory_usage='deep')

```

Análise Univariada com histogramas

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['turns'], kde=True, bins=30, color='royalblue')
plt.title('Distribuição do Número de Turnos por Partida')
plt.xlabel('Quantidade de Turnos')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Análise Bivariada em um Gráfico de Barras e em um Gráfico de Dispersão

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='victory_status', hue='winner', palette='Set2')
plt.title('Status de Vitória por Tipo de Vencedor')
plt.xlabel('Como a partida terminou')
plt.ylabel('Quantidade de Partidas')
plt.legend(title='Vencedor')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='white_rating', y='black_rating', hue='winner', alpha=0.4, palette='vlag')

plt.plot([df['white_rating'].min(), df['white_rating'].max()], 
         [df['white_rating'].min(), df['white_rating'].max()], 
         color='black', linestyle='--', label='Ratings Iguais')

plt.title('Relação entre o Rating das Brancas e das Pretas')
plt.xlabel('Rating das Brancas')
plt.ylabel('Rating das Pretas')
plt.legend(title='Vencedor')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Análise Multivariada Ilustrada em Box-Plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='victory_status', y='white_rating', hue='winner', palette='Pastel1')
plt.title('Distribuição do Rating das Brancas por Status de Vitória e Vencedor')
plt.xlabel('Status da Vitória')
plt.ylabel('Rating do Jogador de Brancas')
plt.legend(title='Vencedor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
